```python
# IQR resale_price
q1 = df["resale_price"].quantile(0.25)
q3 = df["resale_price"].quantile(0.75)

iqr = q3-q1 #Interquartile range
fence_low  = q1-1.5*iqr
fence_high = q3+1.5*iqr
```

```python
# mengecek data outlier resale_price
df.loc[(df["resale_price"] < fence_low) | (df["resale_price"] > fence_high)]
```

```python
# IQR floor_area_sqm
q1 = df["floor_area_sqm"].quantile(0.25)
q3 = df["floor_area_sqm"].quantile(0.75)

iqr = q3-q1 #Interquartile range
fence_low  = q1-1.5*iqr
fence_high = q3+1.5*iqr
```

```python
# mengecek data outlier floor_area_sqm
df.loc[(df["floor_area_sqm"] < fence_low) | (df["floor_area_sqm"] > fence_high)]
```

```python
# fungsi untuk feature month_f
def mon(x):
    if (x == '2019-06') or (x == '2020-06'):
        return 6
    elif (x == '2019-08') or (x == '2020-08') or (x=='2018-08'):
        return 8
    elif (x == '2019-09') or (x=='2020-09'):
        return 9
    elif x == '2019-10':
        return 10
    elif (x == '2019-11') or (x=='2018-11'):
        return 10
    elif (x == '2019-12') or ('2018-12'):
        return 10
    elif (x == '2020-01') or (x=='2019-01'):
        return 1
    elif x == '2020-02':
        return 2
    elif x == '2020-03':
        return 3
    elif (x == '2020-04') or (x=='2019-04'):
        return 4
    elif x == '2020-05':
        return 5
    elif x == '2020-06':
        return 6
    elif x == ('2020-07') or (x=='2018-07'):
        return 7
    
```
